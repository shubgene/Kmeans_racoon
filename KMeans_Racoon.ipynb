```python
import numpy as np
from sklearn import cluster,datasets
from scipy import misc
import scipy as sp
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
face = sp.misc.face(gray = True)
plt.figure(figsize=(10, 10))
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
```

```python
n_clusters = 5
np.random.seed(0)
X=face.shape[0]
Y=face.shape[1]
Z = face.reshape(X*Y,1)
k_Means= cluster.KMeans(n_clusters=n_clusters)
k_Means.fit(Z)
```

```python
clusters = np.asarray(k_Means.cluster_centers_) 
labels = np.asarray(k_Means.labels_)  
labels = labels.reshape(X,-Y); 
 
#np.save('codebook_racoon.npy',clusters)
plt.imsave('compressed_racoon.png',labels);

# Visualize the compressed image

image = plt.imread('compressed_racoon.png')
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.show()
```
